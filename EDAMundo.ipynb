Cambio formato de fecha y genero nuevas columna de decada

```python
df["fecha"] = pd.to_datetime(df["fecha"])
df["fecha"] = df["fecha"].dt.strftime("%d/%m/%Y")
df["fecha"] = pd.to_datetime(df["fecha"])
df['Decada'] = (df['fecha'].dt.year // 10) * 10
df["Decada"] = df["Decada"].dt.strftime("%d/%m/%Y")
df["Decada"] = pd.to_datetime(df["Decada"])
```

Elimino en un principio las columnas que no voy a necesitar

```python
df.drop("HORA declarada",axis="columns",inplace=True)
df.drop("OperadOR",axis="columns",inplace=True)
df.drop("flight_no",axis="columns",inplace=True)
df.drop("route",axis="columns",inplace=True)
df.drop("registration",axis="columns",inplace=True)
df.drop("cn_ln",axis="columns",inplace=True)
df.drop(df.columns[0], axis=1, inplace=True)
```

Cambio los ? por null para poder visualizarlos mejor

```python
df["ac_type"] = df["ac_type"].str.replace("\?", "Desconocido")

df["all_aboard"] = df["all_aboard"].str.replace("\?", "-1")
df["all_aboard"] = df["all_aboard"].replace("-1", np.nan)

df["PASAJEROS A BORDO"] = df["PASAJEROS A BORDO"].str.replace("\?", "-1")
df["PASAJEROS A BORDO"] = df["PASAJEROS A BORDO"].replace("-1", np.nan)

df["crew_aboard"] = df["crew_aboard"].str.replace("\?", "-1")
df["crew_aboard"] = df["crew_aboard"].replace("-1", np.nan)

df["cantidad de fallecidos"] = df["cantidad de fallecidos"].str.replace("\?", "-1")
df["cantidad de fallecidos"] = df["cantidad de fallecidos"].replace("-1", np.nan)

df["passenger_fatalities"] = df["passenger_fatalities"].str.replace("\?", "-1")
df["passenger_fatalities"] = df["passenger_fatalities"].replace("-1", np.nan)

df["crew_fatalities"] = df["crew_fatalities"].str.replace("\?", "-1")
df["crew_fatalities"] = df["crew_fatalities"].replace("-1", np.nan)

df["ground"] = df["ground"].str.replace("\?", "-1")
df["ground"] = df["ground"].replace("-1", np.nan)
```
